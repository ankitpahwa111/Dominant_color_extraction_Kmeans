```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles,make_moons
```

## KMeans clusters on the  basis of Distance , which is not always acceptable

```python
X,Y = make_moons(n_samples=200, noise=0.10)
plt.scatter(X[:,0], X[:,1], c=Y)
plt.show()
```

## The Kmeans of above dataset is :

```python
from sklearn.cluster import KMeans
```

```python
km = KMeans(n_clusters=2)
km.fit(X)
```

```python
centers = km.cluster_centers_
label = km.labels_

plt.scatter(X[:,0], X[:,1], c=label)
plt.scatter(centers[:,0], centers[:,1], color='red')
plt.show()
```

## We can see that Kmeans fucked up , so we need something to cluster based on Density


## DBSCAN
Density Based Spatial Clustering of Applications with Noise
<ul>
    <li>Based upon on the idea, that a cluster is a high density area surrounded by low density region</li>
    <li>Starts by exploring a small area, if density is "good enough" it considered as part of cluster, and explores neigbours and so on, so it groups regions based upon density!</li>
</ul>

DBSCAN Works on the idea of eps , means if a data point has distance less than eps , it is considered part of current cluster , hence the relative distancing is done rather than absolute distancing in Kmeans

```python
from sklearn.cluster import DBSCAN
```

```python
dbs = DBSCAN(eps=0.21, min_samples=5)
```

```python
dbs.fit(X)
```

```python
Ypred = dbs.fit_predict(X)
plt.scatter(X[:,0], X[:,1], c=Ypred)
plt.show()
```
