## KMeans


##  we will make clusters on the data set

Input Data : X = {x1,...,xN}

Goal : Partition the data among some K number of clusters. Let us assume K is known to us.

Let µk denote the center of Kth Cluster (uk will be vector)

So we need to ﬁnd an assignment of data points to clusters, as well as a set of cluster centers{µk}, such that the sum of the squares of the distances of each data point to its closest cluster center µk, is a minimum.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

## Data set Prep

```python

X,Y = make_blobs(n_samples=1000, n_features=2,centers=5,random_state=10)
```

```python
plt.figure(0)
plt.grid(True)
plt.scatter(X[:,0],X[:,1],c = Y)
plt.show()
```

## Defining Clusters

```python
clusters = {}
k = 5 # 5 clusters
colors = ['green', 'red', 'blue', 'yellow', 'gray']

# Each Cluster will have 3 things : Center Point, List Of points belonging to that cluster and color of the cluster

for i in range(k):
    center = 10*(2*np.random.random((X.shape[1],))-1)
    points = []
    cluster = {
        'center':center, 
        'points':points,
        'color':colors[i]
    }
    clusters[i] = cluster
```

## Making clusters

```python
def distance(x1,x2):
    return np.sqrt(np.sum((x1-x2)**2))
```

```python
# Assigning points based on distance to cluster center
def AssignPointsToCluster(clusters):
    for ix in range(X.shape[0]):
        
        dist = []
        for kx in range(k):
            d = distance(X[ix], clusters[kx]['center'])
            dist.append(d)
            
        current_cluster = np.argmin(dist)
        clusters[current_cluster]['points'].append(X[ix])
```

```python
# Updating Clusters
def updateClusters(clusters):
    for kx in range(k):
        pts = np.array(clusters[kx]['points'])
        
        if pts.shape[0]>0: # if cluster has some non-zero points
            new_mean = pts.mean(axis=0)
            clusters[kx]['center'] = new_mean
            #clear my points list 
            clusters[kx]['points'] = []      
        
```

```python
# plot clusters
def plotClusters(clusters):
    for kx in range(k):
        
        pts = np.array(clusters[kx]['points'])
        
        try:
            plt.scatter(pts[:,0], pts[:,1], color=clusters[kx]['color'])
        except:
            pass
        
        # plot the cluster center
        uk = clusters[kx]['center']
        plt.scatter(uk[0], uk[1], color='black', marker='*')
        
```

```python
# Repeat the process of step 2 and step 3 until convergence 

count =1 
while True:
    AssignPointsToCluster(clusters)
    updateClusters(clusters)
    count+=1
    if count>10:
        break
```

```python
AssignPointsToCluster(clusters)
plotClusters(clusters)
updateClusters(clusters)
```
